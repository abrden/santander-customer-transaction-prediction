```python
import numpy as np
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split

# NOTE: Make sure that the class is labeled 'target' in the data file
tpot_data = pd.read_feather('data_feather.fth')
ids = tpot_data['ID_code']
target = tpot_data['target']
tpot_data.drop(columns=['ID_code'], inplace=True)
tpot_data.drop(columns=['target'], inplace=True)

training_features, testing_features, training_target, testing_target = \
            train_test_split(tpot_data, target.values, random_state=42)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Average CV score on the training set was:0.914499999377013
exported_pipeline = LogisticRegression(C=0.0001, dual=True, penalty="l2")

exported_pipeline.fit(training_features, training_target)
results = exported_pipeline.predict(testing_features)
```
