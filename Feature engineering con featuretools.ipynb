```python
import pandas as pd
data = pd.read_csv("train.csv")
target = data['target']
```

```python
import featuretools as ft

entityset = ft.EntitySet()
entityset = entityset.entity_from_dataframe(entity_id = 'transactions', dataframe = data, index = 'ID_code')

# Perform deep feature synthesis without specifying primitives
features, feature_names = ft.dfs(entityset=entityset, target_entity='transactions', max_depth = 15)
```
