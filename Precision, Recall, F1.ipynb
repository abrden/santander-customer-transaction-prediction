```python
import numpy as np
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split

# NOTE: Make sure that the class is labeled 'target' in the data file
tpot_data = pd.read_csv('train_trimmed.csv')
ids = tpot_data['ID_code']
target = tpot_data['target']
tpot_data.drop(columns=['ID_code'], inplace=True)
tpot_data.drop(columns=['target'], inplace=True)

training_features, testing_features, training_target, testing_target = \
            train_test_split(tpot_data, target.values, random_state=42)
```

```python
# Replace this cell with your model. Save the target predictions in an array called 'results'
exported_pipeline = GradientBoostingClassifier(learning_rate=0.5, max_depth=4, max_features=0.05, min_samples_leaf=8, min_samples_split=12, n_estimators=100, subsample=0.9500000000000001)

exported_pipeline.fit(training_features, training_target)
results = exported_pipeline.predict(testing_features)
```

```python
results
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
precision_score(testing_target, results) #, average='micro'
```
